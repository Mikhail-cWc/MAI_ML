```python
from sklearn.datasets import make_regression
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

df = make_regression()
model = Ridge()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[0], df[1], test_size=0.2)
```

```python
model.fit(x_train, y_train)
y_train_pred = model.predict(x_train)
y_pred = model.predict(x_test)

print(f"MSE loss: {mean_squared_error(y_train, y_train_pred, squared = False)}")
print(f"MSE loss: {mean_squared_error(y_test, y_pred, squared = False)}")
```
